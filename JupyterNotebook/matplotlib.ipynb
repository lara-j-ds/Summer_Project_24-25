```python
# Create a figure
fig = plt.figure(figsize=(10, 6))
# Ask, out of a 1x1 grid of plots, the first axes.
ax = fig.add_subplot(1, 1, 1)
# Plot times as x-variable and temperatures as y-variable
ax.plot(times, temps);
# Add some labels to the plot
ax.set_xlabel('Time')
ax.set_ylabel('Temperature')
ax.set_title('GFS Temperature Forecast', size=16)
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1)

# Plot two series of data
# The label argument is used when generating a legend.
ax.plot(times, temps, label='Temperature (surface)')
ax.plot(times, temps_1000, label='Temperature (1000 mb)')

# Add labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Temperature')
ax.set_title('Temperature Forecast')

# Add gridlines
ax.grid(True)

# Add a legend to the upper left corner of the plot
ax.legend(loc='upper left');
```

```python
x = y = np.arange(-3.0, 3.0, 0.025)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-(X**2) - Y**2)
Z2 = np.exp(-((X - 2) ** 2) - (Y - 1) ** 2)
Z = (Z1 - Z2) * 2
fig, ax = plt.subplots(figsize=(8,10))
im = ax.imshow(
    Z, interpolation='bilinear', cmap='PiYG', origin='lower', extent=[-3, 3, -3, 3]
)
c = ax.contour(X,Y,Z, levels=np.arange(-2, 2, 0.25), cmap='twilight')
ax.clabel(c);
```

```python
from matplotlib.animation import FuncAnimation
# pacman open/close in place
colors = ["white", "yellow"]
nums = [0,1]
eye_n = [0.9, 0.1]
eye_c = ["black","yellow"]
fig, ax = plt.subplots()
fig.suptitle("test")

def update(num):
    ax.clear()
    if(num>=10):
        nums[0] -= 0.01
        nums[1] += 0.02
    else:
        nums[0] += 0.01
        nums[1] -= 0.02

    x = ax.pie(nums, colors=colors, normalize=False)
    eye = ax.pie(eye_n, colors=eye_c, startangle=45, radius=0.15, center=[0.25,0.5])
    z = ax.pie([0,1], colors=['w', 'w'], center=[0.001,0.001], radius=0.01)

anim = FuncAnimation(fig, update, frames=20, interval=200)
anim

anim.save('pacman.gif', writer='pillow');
```

```python
from matplotlib.animation import FuncAnimation
## pacman open/close while moving b1
colors = ["white", "yellow", "white"]
nums = [0,40,0]
eye_n = [0.9, 0.1]
eye_c = ["black","yellow"]
fig, ax = plt.subplots(figsize=[10,10])
# fig.suptitle("PAC-MAN")
num=20
def update(num):
    ax.clear()

    if(0<num<10 or 20<num<30):
        nums[0] += 0.5
        nums[2] += 0.5
    if(10<num<20 or 30<num<40):
        nums[0] -= 0.5
        nums[2] -= 0.5

    pos = -2 + num*0.1
    x = ax.pie(nums, colors=colors, normalize=True, center = (pos,0))
    eye = ax.pie(eye_n, colors=eye_c, startangle=45, radius=0.15, center=[0.25 + pos,0.5])
    z = ax.pie([1], colors=['w'], center=[0,0], radius=0.0001)

anim = FuncAnimation(fig, update, frames=40, interval=75)
anim

anim.save('pacman-b1.gif', writer='pillow');
```

```python
from matplotlib.animation import FuncAnimation
## pacman open/close while moving b2
colors = ["white", "yellow", "white"]
nums = [5,40,5]
eye_n = [0.9, 0.1]
eye_c = ["black","yellow"]
fig, ax = plt.subplots(figsize=[10,10])
# fig.suptitle("PAC-MAN")
num=20
def update(num, start_pos=0):
    ax.clear()
    if(num>=20):
        start_pos=-4
    if(0<num<10 or 20<num<30):
        nums[0] -= 0.5
        nums[2] -= 0.5
    if(10<num<20 or 30<num<40):
        nums[0] += 0.5
        nums[2] += 0.5

    pos = start_pos + num*0.1
    x = ax.pie(nums, colors=colors, normalize=True, center = (pos,0))
    eye = ax.pie(eye_n, colors=eye_c, startangle=45, radius=0.15, center=[0.25 + pos,0.5])
    z = ax.pie([1], colors=['w'], center=[0,0], radius=0.0001)

anim = FuncAnimation(fig, update, frames=40, interval=75)
anim

anim.save('pacman-b2.gif', writer='pillow');
```
